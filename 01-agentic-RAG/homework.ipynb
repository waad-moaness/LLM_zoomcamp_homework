```python
from minsearch import Index
import requests
```

```python
def build_index(documents):
    index = Index(
        text_fields=['content'],
        keyword_fields=['filename']
    )
    index.fit(documents)
    return index
```

```python
index= build_index(documents)
```

```python
query = 'How does the agentic loop keep calling the model until it stops?'

result = index.search(query)
```
